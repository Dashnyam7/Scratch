# What is logistic regression?

This algorithm is mainly used for survival analysis (live or die).
Survival analysis is applied not only in the fields of disease and medicine, but also in a broader sense, to problems such as whether or not a user will take a specific action, and is used in various fields and fields.

In short, it is an algorithm that calculates the probability of an event occurring .

### Explanation of mathematical aspects

Logistic regression is a classification model that follows Bernoulli distribution or binomial distribution and is a kind of general linear model .

First, in the general linear model, the value obtained by multiplying the variables by an arbitrary coefficient and adding them is called the linear combination of the variables, like in linear regression.
This linearly combined value is "transformed by a function" to obtain the final prediction result.

The output of logistic regression is the "probability of an event occurring", and class classification is performed from this output result.

### Assumptions of Logistic Regression

1. The explanatory variable (x) is continuous or discrete, and the objective variable (y) is discrete.
2. The output value of logistic regression is a continuous value between 0 and 1
3. The final predicted value follows a 0 or 1 Bernoulli/binomial distribution (mean np, variance np(1-p))

### What is a regularization term?

Machine learning algorithms learn by using training data to improve prediction accuracy for unknown data. During learning, over-learning the training data may reduce the prediction accuracy for unknown data. It can be considered that the model over-trained on the training data and learned outliers and noise. This phenomenon is called overfitting , overfitting , or overfitting .

A regularization term is an item that can be introduced into the loss function (objective function) to suppress this overfitting.

There are two types of regularization terms: L1 regularization and L2 regularization. L1 regularization uses the sum of the absolute values of the parameters, and L2 regularization uses the sum of the squares of the parameters as the penalty term. L1 regularization has a strong implication of omitting redundant explanatory variables (features), and L2 regularization has a strong implication of suppressing overfitting.

```python
def _gradient_descent(self, X, y):
    m = X.shape[0]
    n = X.shape[1]
    pred = self._logistic_hypothesis(X)
    for j in range(n):
        gradient = 0
        for i in range(m):
            gradient += (pred[i] - y[i]) * X[i, j]
        self.theta[j] = self.theta[j] - self.lr * ((gradient+self.lam*self.theta[j]) / m)
```

### create function

```python
def _loss_func(self, pred, y):
    error = 0
    for i in range(y.shape[0]):
        error += -np.sum(y[i] *  np.log(pred[i])+(1-y[i]) *  np.log(1-pred[i]))
    loss = error / (y.shape[0])
    loss = loss + np.sum(self.theta**2)*self.lam/(2 * y.shape[0])
    return loss
```

```python
def fit(self, X, y, X_val, y_val): 
    if self.bias == True:
      bias = np.ones((X.shape[0], 1))
      X = np.hstack((bias, X))
      bias = np.ones((X_val.shape[0], 1))
      X_val = np.hstack((bias, X_val))
    self.theta = np.zeros(X.shape[1])
    self.theta = self.theta.reshape(X.shape[1], 1)
    for i in range(self.num_iter):
        pred = self._logistic_hypothesis(X)
        pred_val = self._logistic_hypothesis(X_val)
        self._gradient_descent(X, y)
        loss = self._loss_func(pred, y)
        self.loss = np.append(self.loss, loss)
        loss_val = self._loss_func(pred_val, y_val)
        self.val_loss = np.append(self.val_loss, loss_val)
        if verbose == True:
            print('{}{}'.format(i,loss))
```

```python
class ScratchLogisticRegression():
    def __init__(self, num_iter, lr, bias, verbose,lam):
        self.num_iter = num_iter
        self.lr = lr
        self.bias = bias
        self.verbose = verbose
        self.lam = lam
        self.theta = np.array([])
        self.loss = np.array([])
        self.val_loss = np.array([])

    def fit(self, X, y, X_val=None, y_val=None):
        if self.bias == True:
            bias = np.ones((X.shape[0], 1))
            X = np.hstack((bias, X))
            bias = np.ones((X_val.shape[0], 1))
            X_val = np.hstack((bias, X_val))
        self.theta = np.zeros(X.shape[1])
        self.theta = self.theta.reshape(X.shape[1], 1)
        for i in range(self.num_iter):
            pred = self._logistic_hypothesis(X)
            pred_val = self._logistic_hypothesis(X_val)
            self._gradient_descent(X, y)
            loss = self._loss_func(pred, y)
            self.loss = np.append(self.loss, loss)
            loss_val = self._loss_func(pred_val, y_val)
            self.val_loss = np.append(self.val_loss, loss_val)
            if self.verbose == True:
                print('{}th training loss is {}'.format(i,loss))

    def _gradient_descent(self, X, y):
        m = X.shape[0]
        n = X.shape[1]
        pred = self._logistic_hypothesis(X)
        for j in range(n):
            gradient = 0
            for i in range(m):
                gradient += (pred[i] - y[i]) * X[i, j]
            self.theta[j] = self.theta[j] - self.lr * ((gradient+self.lam*self.theta[j]) / m)
     
    def _sigmoid(self,y):
        """sigmoid function"""
        return 1 / (1 + np.exp(-y))
        
    def _logistic_hypothesis(self, X):
        pred = X @ self.theta
        pred = self._sigmoid(pred)
        return pred
       
    def predict(self, X):
        """
        Estimate labels using logistic regression

        Parameters
        ----------
        X : ndarray, shape (n_samples, n_features)
            sample

        Returns
        -------
            ndarray, shape (n_samples, 1)
            Estimation result by logistic regression
        """

        if self.bias == True:
            a = np.ones(X.shape[0]).reshape(X.shape[0], 1)
            X = np.hstack([a, X])
        return  np.where(self._logistic_hypothesis(X) >= 0.5,1,0)

    def predict_proba(self, X):
        """
        Estimate probabilities using logistic regression

        Parameters
        ----------
        X : ndarray, shape (n_samples, n_features)
            sample

        Returns
        -------
            ndarray, shape (n_samples, 1)
            Estimation result by logistic regression
        """

        if self.bias == True:
            a = np.ones(X.shape[0]).reshape(X.shape[0], 1)
            X = np.hstack([a, X])
        pred = self._logistic_hypothesis(X)
        return pred
        
    def _loss_func(self, pred, y):
        error = 0
        for i in range(y.shape[0]):
            error += -np.sum(y[i] *  np.log(pred[i])+(1-y[i]) *  np.log(1-pred[i]))
        loss = error / (y.shape[0])
        loss = loss + np.sum(self.theta**2)*self.lam/(2 * y.shape[0])
        return loss
```

```python
from matplotlib.colors import ListedColormap
def decision_region(X,y,slr):
    mesh_f0, mesh_f1  = np.meshgrid(
        np.arange(np.min(X[:,0]), np.max(X[:,0]), 0.01), 
        np.arange(np.min(X[:,1]), np.max(X[:,1]), 0.01)
    )
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = slr.predict(mesh).reshape(mesh_f0.shape)
    plt.title('decision region')
    plt.xlabel('feature0')
    plt.ylabel('feature1')
    plt.contourf(mesh_f0, mesh_f1, y_pred,cmap=ListedColormap(['pink', 'skyblue']))
    plt.contour(mesh_f0, mesh_f1, y_pred,colors='red')
    plt.scatter(X[y==0][:, 0], X[y==0][:, 1],label='0')
    plt.scatter(X[y==1][:, 0], X[y==1][:, 1],label='1')
    plt.legend()
    plt.show()
```
